```python
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv("C:/Users/oluwa/DATA SCIENCE/Data Science & ML/Reinforcing my ML Knowledge/sonar_dataset.csv", header=None) #"header=None" uses an integer
                                                                                                                        #sequence to automatically
                                                                                                                        #name each column is the columns
                                                                                                                        #have no headings
df.sample(5)
```

```python
X  = df.drop(60, axis=1)
y = df[60]
y.head()
```

```python
y = pd.get_dummies(y, drop_first=True) #this line of code takes a categorical Series y and 
                                       #transforms it into a set of binary (0 or 1) columns, where each column represents a category.
y = y.astype(int)
y.sample(5)    #takes five random samples out of the newly created column
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
X_train.shape, X_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'), #we are using 60 neurons because we have 60 columns in our training set
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', #because it is a binary classification problem
              optimizer='adam', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test).reshape(-1)
print(y_pred[:10])

#round the values to the nearest integer, that is 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test, y_pred))
```

## Now we want to try a model with a dropout layer

```python
modeld = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'), 
    keras.layers.Dropout(0.5), #We introduced a dropout layer to drop 50% of the neurons
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dropout(0.5), #The normal things is to introduce a dropout layer after a hidden layer
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation='sigmoid')
])

modeld.compile(loss='binary_crossentropy', #because it is a binary classification problem
              optimizer='adam', metrics=['accuracy'])
modeld.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
modeld.evaluate(X_test, y_test)
```

## Now, notice that the training accuracy has reduced to about 83%, but the test set accuracy has also increased to about 78%, meaning that the dropout layer actually did its work well, reducing overfitting in the training set

```python
y_pred = modeld.predict(X_test).reshape(-1)
print(y_pred[:10])

#round the values to the nearest integer, that is 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test, y_pred))
```

# In summary, dropout regularization is more like a trial and error method, but it works most time, particularly in Computer Vision, where you can have a lot of complex neural networks.
